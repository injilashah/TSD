```python
import pandas as pd

data = pd.read_csv("/content/natural_gas_price.csv",index_col="Date",parse_dates=True)
data.head(20)
```

```python
duplicates = data[data.index.duplicated()]
duplicates
```

```python
data.dropna(inplace = True)
```

```python
print("Inferred frequency:", data.index.freq)
```

lets calculate missing dates

```python


full_range = pd.date_range(start=data.index.min(), end=data.index.max(), freq='D')
missing_dates = full_range.difference(data.index)

print("Missing days:", len(missing_dates))
print(missing_dates)
```
